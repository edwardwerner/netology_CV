## Визуализация в matplotlib

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

Изображение выглядит странновато. Это потому, что красный, синий и зелёный стоят не в том порядке, который ожидает матплотлиб.

```python
# изменяем порядок каналов для визуализации (b, g, r) -> (r, g, b)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

Самый первый пиксель:

```python
img[0][0]
```

```python
img2[0][0]
```

Все красные пиксели:

```python
img2[:, :, 0]
```

## Переходим в пространство цветов HSV

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(1, figsize = (16, 4))
plt.subplot(141)
plt.imshow(img2)
plt.subplot(142)
plt.imshow(img_hsv[...,0], cmap='gray')
plt.title('Hue')
plt.subplot(143)
plt.imshow(img_hsv[...,1], cmap='gray')
plt.title('Saturation')
plt.subplot(144)
plt.imshow(img_hsv[...,2], cmap='gray')
plt.title('Value');
```

#### Справка про numpy
`matrix[..., 0]` -- в данном случае то же, что `matrix[:, :, 0]`: взять второй элемент из самых вложенных массивов.

```python
img_hsv[..., 0]
```

```python
img_hsv[:, :, 0]
```
